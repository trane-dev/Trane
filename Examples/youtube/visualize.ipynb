```python
import json
import numpy as np
```

```python
with open("channel_title_result.json") as f:
    results = json.load(f)

classifications = list(filter(lambda x: x["problem_type"] == "classification", results))
regressions = list(filter(lambda x: x["problem_type"] == "regression", results))
```

```python
counter = 0
for item in classifications + regressions:
    if len(item["evaluations"]) == 0:
        print(item)
        counter += 1
print(counter)
```

```python
def temp_and_score_cl(item):
    return {
        "nl": item["template_nl"],
        "score": -1
        if len(item["evaluations"]) == 0
        else np.mean(
            [np.max(list(x["Accuracy"].values())) for x in item["evaluations"]]
        ),
    }


classifications_sorted = [temp_and_score_cl(item) for item in classifications]


def temp_and_score_re(item):
    return {
        "nl": item["template_nl"],
        "score": -1
        if len(item["evaluations"]) == 0
        else np.mean([np.max(list(x["R2"].values())) for x in item["evaluations"]]),
    }


regressions_sorted = [temp_and_score_re(item) for item in regressions]
```

```python
classifications_sorted = sorted(classifications_sorted, key=lambda x: -x["score"])
regressions_sorted = sorted(regressions_sorted, key=lambda x: -x["score"])
```
