```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("flights.csv", dtype={"TAIL_NUMBER":str, "FLIGHT_NUMBER":str, "ORIGIN_AIRPORT": str, "DESTINATION_AIRPORT": str})
# df_head = df.head(100)
```

```python
# df = df_head.copy()
df.head()
```

```python
df['DATE'] = df.apply(lambda row: "%04d-%02d-%02d" % (row['YEAR'], row['MONTH'], row["DAY"]), axis=1)
```

```python
df.head()
```

```python
df['SCHEDULED_DEPARTURE_HOUR'] = df.apply(lambda row: row['SCHEDULED_DEPARTURE']//100, axis=1)
df['SCHEDULED_DEPARTURE_MINUTE'] = df.apply(lambda row: row['SCHEDULED_DEPARTURE']%100, axis=1)
```

```python
df.head()
```

```python
df['CANCELLATION_REASON'] = df['CANCELLATION_REASON'].apply(lambda x: '-' if not isinstance(x, str) else x)
df['TAIL_NUMBER'] = df['TAIL_NUMBER'].apply(lambda x: '-' if not isinstance(x, str) else x)
df['FLIGHT_NUMBER'] = df['AIRLINE'] + df['FLIGHT_NUMBER']
```

```python
df.head()
```

```python
for item in ['ELAPSED_TIME', 'DEPARTURE_DELAY', 'ARRIVAL_DELAY', 'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']:
    df[item] = df[item].apply(lambda x: 0 if np.isnan(x) else x)
```

```python
new_df = df[['DATE', 'DAY_OF_WEEK', 'AIRLINE', 'FLIGHT_NUMBER', 'TAIL_NUMBER', 
    'ORIGIN_AIRPORT', 'DESTINATION_AIRPORT', 'SCHEDULED_DEPARTURE_HOUR', 
    'SCHEDULED_TIME', 'ELAPSED_TIME', 'DEPARTURE_DELAY', 'ARRIVAL_DELAY', 
    'CANCELLED', 'CANCELLATION_REASON', 
    'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']]
new_df.to_csv("flight-clean.csv", index=None)
```
