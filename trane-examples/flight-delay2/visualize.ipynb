```python
import json
import numpy as np
```

```python
with open("prob_with_acc.json") as f:
    results = json.load(f)

classifications = list(filter(lambda x: x['problem_type'] == 'classification', results))
regressions = list(filter(lambda x: x['problem_type'] == 'regression', results))
```

```python
counter = 0
for item in classifications:
    if len(item['evaluations']) == 0:
        print(item)
        counter += 1
print(counter)
```

```python
def temp_and_score(item):
    return {
        "nl": item['template_nl'], 
        "score": -1 if len(item['evaluations']) == 0 else np.mean([np.max(list(x['Accuracy'].values())) for x in item['evaluations']])
    }

classifications_sorted = [temp_and_score(item) for item in classifications]
```

```python
classifications_sorted = sorted(classifications_sorted, key=lambda x: -x['score'])
```
